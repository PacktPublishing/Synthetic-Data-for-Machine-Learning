```python
# Chapter 4 "Synthetic Data for Machine Learning"

# A program to show how to use random numbers generator to generate synthetic data from real data

import random
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')


num_days = 11
synthetic_data = []# list to store the synthetic data
synth_real_approx_factor = 0.5# 0 prefect synthetic data; 1 very noisy synthetic data

# Data range
range_min = -10
range_max =  10

# Our independent attribute
days  = np.arange(0, num_days, 0.1);

# Magentidue of sin wave is set to range_max
real_data   = np.sin(days)*range_max

# Generate synthetic data
for i in range(0,len(days)):
  synthetic_data_sample = real_data[i] + (random.random()-0.5)*range_max*synth_real_approx_factor

  synthetic_data.append(synthetic_data_sample)

# Plot real data
plt.plot(days, real_data,label='Real Data')

# Plot synthetic data
plt.plot(days, synthetic_data, 'o', color='black', label='Synthetic Data');

# add figure's axes labels, main title, and legend 
plt.title('Real and Synthetic Data')

plt.xlabel('Time (Days)')
plt.ylabel('Profit (GBP)')

plt.grid(True, which='both')

plt.axhline(y=0, color='k')

plt.legend(loc="lower left")

# Show the plot
plt.show()
```
